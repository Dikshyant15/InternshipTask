```python
import requests
from bs4 import BeautifulSoup
import json
```

```python
with open(r"D:\Internship\EnterpriseKnowledgeBased\data\hrefs_dict2.json") as f:
    links_data=json.loads(f.read())
```

```python
list1=links_data.get("icon_accounts\rAccountsicon_arrow-front")
type(list1)
```

```python
list1
```

```python
rfcd_data={}
for parse_link in list1:
    rfcd_data[parse_link]={}
    response=requests.get(parse_link)
    data=response.text
    try:
        soup=BeautifulSoup(data,"html.parser")
        container=soup.find(name="div",attrs={"class": "question-container items"})
        items=container.findAll(name="div",attrs={"class":"item"})
        for a in items:
            rfcd_data[parse_link][a.a.text]=[]
            rfcd_data[parse_link][a.a.text]=[i.text for i in a.findAll("li")]
    except Exception as e:
        container=soup.find(name="div",attrs={"class": "container"})
        rfcd_data[parse_link]=container.getText()
        
        

rfcd_data
```

```python
import json
import os

# Specify the directory and filename
directory = 'D:\Internship\EnterpriseKnowledgeBased\data'  # Change this to your designated path
filename = 'rfcd_data.json'
file_path = os.path.join(directory, filename)

# Ensure the directory exists
os.makedirs(r'D:\Internship\EnterpriseKnowledgeBased\data', exist_ok=True)

# Write the JSON data to the file
with open(file_path, 'w') as f:
    json.dump(rfcd_data, f, indent=4)  # Using json.dump for better readability
```
