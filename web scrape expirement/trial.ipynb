```python
import requests
from bs4 import BeautifulSoup
import pandas as pd


response=requests.get("https://citybankplc.com/obu")

data=response.text
```

```python
soup = BeautifulSoup(data,"html.parser")
soup
```

```python
faq = soup.find(name="div",attrs={"class":"question-container items"})
faq
```

```python
obu = faq.findAll(name="div", attrs={"class":"item"})
obu
```

```python
obu_faq = obu[2]
obu_faq
```

```python
card_data_obu = obu_faq.findAll(name = "div", attrs={"class":"carditem"})
card_data_obu
```

```python
ib_faq = obu[3]
ib_faq
```

```python
card_data_ib= ib_faq.findAll(name = "div", attrs={"class":"carditem"})
card_data_ib
```

```python
faq_data = []
current_question = None
# questions = [input.text.strip() for input in input_sets if input.text.strip()[:1].isdigit()]

# Loop through all paragraph and list items within the div
for card in card_data_obu:
    for tag in card.find_all(["p", "li"]):
        if tag.strong and tag.strong.text.strip()[:1].isdigit():  # This is a question
            if current_question:  # Save the previous question and its answers before moving to the next
                faq_data.append(current_question)
            # Start a new question dictionary
            current_question = {"question": tag.get_text(strip=True), "answers": []}
        elif current_question:  # This is an answer related to the current question
            current_question["answers"].append(tag.get_text(strip=True))
            
for card in card_data_ib:        
    for tag in card.find_all(["p", "li"]):
        if tag.strong and tag.strong.text.strip()[:1].isdigit():  # This is a question
            if current_question:  # Save the previous question and its answers before moving to the next
                faq_data.append(current_question)
            # Start a new question dictionary
            current_question = {"question": tag.get_text(strip=True), "answers": []}
        elif current_question:  # This is an answer related to the current question
            current_question["answers"].append(tag.get_text(strip=True))
# Append the last question if there's any
if current_question:
    faq_data.append(current_question)
```

```python
# Print the structured data
for faq in faq_data:
    print(f"Question: {faq['question']}")
    for answer in faq['answers']:
        print(f"Answer: {answer}")
    print('-' * 30)
```

```python
# Initialize a list to hold all Q&A dictionaries
all_qa = []
# Find all div elements that have ids starting with 'Accountopen'
account_divs = soup.find_all('div', id=lambda x: x and (x.startswith('Accountopen') or x.startswith('ibAccountopen')))

for div in account_divs:
    qa_dict = {'question': [], 'answer': []}
    elements = div.find_all(['p', 'ul'])  # Find all <p> and <ul> tags within the div

    for element in elements:
        if element.name == 'p' and element.strong:
            # This is a question
            question_text = element.get_text(strip=True)
            qa_dict['question'].append(question_text)
        elif element.name == 'p':
            # This is a continuation of an answer or additional info
            answer_text = element.get_text(strip=True)
            if qa_dict['question']:
                qa_dict['answer'].append(answer_text)  # Add answer text to the last found question
        elif element.name == 'ul':
            # This is a list within an answer
            list_items = element.find_all('li')
            list_text = ' '.join(li.get_text(strip=True) for li in list_items)
            if qa_dict['question']:
                qa_dict['answer'].append(list_text)  # Add list text to the last found question

    # Store the compiled Q&A in the main list if there's any content
    if qa_dict['question']:
        all_qa.append(qa_dict)

# Output the structured data
for qa in all_qa:
    print(f"Questions: {qa['question']}")
    print(f"Answers: {qa['answer']}")
    print('-' * 50)
```

```python
faq_df = pd.DataFrame(faq_data)
```
