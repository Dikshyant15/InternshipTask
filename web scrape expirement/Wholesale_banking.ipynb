```python
wholesale_banking_data = {}
links2 = ['https://www.citybankplc.com/working-capital-finance','https://www.citybankplc.com/shortmid-term-finance',
          'https://www.citybankplc.com/structured-finance','https://www.citybankplc.com/investment-banking']
        #   'https://www.citybankplc.com/structured-finance'

#WORKING CAPITAL FINANCE
working_capital_finanace = list1[0]
working_capital_finanace_response = requests.get(working_capital_finanace)
working_capital_finanace_data = working_capital_finanace_response.text
working_capital_finanace_soup = BeautifulSoup(working_capital_finanace_data,"html.parser")
# print(working_capital_finanace_soup)

#TRADE FINANCE 
trade_finanace = list1[1]
trade_finanace_response = requests.get(working_capital_finanace)
trade_finanace_data = trade_finanace_response.text
trade_finanace_finanace_soup = BeautifulSoup(trade_finanace_data,"html.parser")

#SHORT/MID TERM FINANCE
sm_finance = list1[2]
sm_finance_response = requests.get(sm_finance)
sm_finance__data = sm_finance_response = requests.get(sm_finance).text
sm_finance__soup = BeautifulSoup(working_capital_finanace_data,"html.parser")

#PROJECT FINANCE
project_finance = list1[3]
project_finance_response = requests.get(project_finance)
project_finance_data = requests.get(project_finance).text
project_finance_soup = BeautifulSoup(project_finance_data,"html.parser")

#STRUCTURED FINANCE
structured_finanace = list1[4]
structured_finanace_response = requests.get(structured_finanace)
structured_finanace_data = structured_finanace_response.text
structured_finanace_soup = BeautifulSoup(structured_finanace_data,"html.parser")

#CASH MANAGEMENT
cash_management = list1[6]
cash_management_response = requests.get(cash_management)
cash_management_data = cash_management_response.text
cash_management_response_soup = BeautifulSoup(cash_management_data,"html.parser")


for link in links2:
    response = requests.get(link)
    data = response.text
    try:
        #simliar structure of the links2
        soup = BeautifulSoup(data, "html.parser")
        containers = soup.find(name="div", attrs={'class': "container"})
        # print(containers)
        content_dict = {}
        if containers:
            # Find all child divs within the container
            child_divs = containers.find_all(name="div", recursive=False)
            for child_div in child_divs:
                h4_tags = child_div.find_all('h4')
                p_tags = child_div.find_all('p')
                # print(p_tags)
                for h4_tag, p_tag in zip(h4_tags, p_tags):
                    content_dict[h4_tag.text.strip()] = p_tag.text.strip() 
        # Store the extracted content in the dictionary under the specific key
        wholesale_banking_data[link] = content_dict
    except Exception as e:
        print(f"An error occurred: {e}")

#Trade finance data 
response = requests.get(trade_finanace)
data = response.text
try:
    soup = BeautifulSoup(data,"html.parser")
    containers= soup.find_all(name="div",attrs={"class":"question-container items"})

    all_trade_finance_data = {}
    
    for container in containers:
        # Initialize a dictionary for each container
        trade_finance_dict = {}
        
        # Find all items within the container
        items = container.find_all(name='div', attrs={"class": "item"})
        
        for item in items:
            # Extract the key from the <a> tag
            key = item.a.text.strip()
            
            # Extract the paragraphs (<p> tags) and store them as a list
            paragraphs = [i.text.strip() for i in item.find_all("p")]
            
            # Store the key-value pair in the dictionary
            trade_finance_dict[key] = paragraphs
        
        # Add this container's data to the main dictionary
        all_trade_finance_data.update(trade_finance_dict)
    
    # Store the aggregated trade finance data in the main dictionary
    wholesale_banking_data[trade_finanace] = all_trade_finance_data
except Exception as e:
    container=soup.find(name="div",attrs={"class": "container"})
    wholesale_banking_data[trade_finanace]=container.getText()

#Project Finance 
response = requests.get(project_finance)
data = response.text
try:
    soup = BeautifulSoup(data,"html.parser")
    container = soup.find(name="div", attrs={"class": "container"})
    project_financing_data = {}

    # Extract the h4 and corresponding content
    h4_tags = container.find_all('h4')
    for h4_tag in h4_tags:
        key = h4_tag.text.strip()
        # If the next sibling is a <ul> or <p>, get its content
        if h4_tag.find_next_sibling('ul'):
            value = [li.text.strip() for li in h4_tag.find_next_sibling('ul').find_all('li')]
        elif h4_tag.find_next_sibling('p'):
            value = h4_tag.find_next_sibling('p').text.strip()
        project_financing_data[key] = value
    wholesale_banking_data[project_finance] = project_financing_data

    # Extract the additional p tag (if any)
    p_tags = container.find_all('p')
    for p_tag in p_tags:
        if p_tag.find('a'):
            project_financing_data['Project Finance Link'] = p_tag.find('a')['href']
except Exception as e:
    print(f"An error occurred: {e}")

#Structured finance data
response = requests.get(structured_finanace)
data = response.text
try:
        # Parse the HTML content
    soup = BeautifulSoup(data, "html.parser")
    container = soup.find(name="div", attrs={'class': "container"})
    content_dict = {}
        
    if container:
            # Find all child divs within the container
        child_divs = container.find_all(name="div", recursive=False)
        for child_div in child_divs:
            h4_tags = child_div.find_all('h4')
                
            for h4_tag in h4_tags:
                key = h4_tag.text.strip()
                content_list = []

                    # Find the next siblings of the h4 tag
                for sibling in h4_tag.find_next_siblings():
                    if sibling.name == 'h4':
                        break
                    if sibling.name == 'p':
                        content_list.append(sibling.text.strip())
                    if sibling.name == 'li':
                        content_list.append(sibling.text.strip())

                content_dict[key] = content_list

            # Store the extracted content in the dictionary under the specific key
        wholesale_banking_data[structured_finanace] = content_dict

except Exception as e:
    print(f"An error occurred: {e}")

#cash management 
response = requests.get(cash_management)
data = response.text
try:
    # Parse the HTML content
    soup = BeautifulSoup(data, "html.parser")
    container = soup.find(name="div", attrs={'class': "container"})
    content_dict = {}
        
    if container:
        # Find all child divs within the container
        child_divs = container.find_all(name="div", recursive=False)
        for child_div in child_divs:
            h4_tags = child_div.find_all('h4')
                
            for h4_tag in h4_tags:
                key = h4_tag.text.strip()
                content_list = []

                    # Find the next siblings of the h4 tag
                for sibling in h4_tag.find_next_siblings():
                    if sibling.name == 'h4':
                        break
                    if sibling.name == 'p':
                        content_list.append(sibling.text.strip())
                    if sibling.name == 'li':
                        content_list.append(sibling.text.strip())

                content_dict[key] = content_list

            # Store the extracted content in the dictionary under the specific key
        wholesale_banking_data[cash_management] = content_dict


except Exception as e:
    print(f"An error occurred: {e}")


wholesale_banking_data
```
