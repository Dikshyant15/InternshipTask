```python
import requests
from bs4 import BeautifulSoup
import json
```

```python
with open(r"D:\Internship\EnterpriseKnowledgeBased\data\hrefs_dict2.json") as f:
    links_data=json.loads(f.read())
```

```python
agent_banking_data={}

for link in list1:
    response = requests.get(link)
    data = response.text
    try:
        soup = BeautifulSoup(data,"html.parser")
        container= soup.find(name="div",attrs={"class":"question-container items"})
        items = container.find_all(name='div',attrs={"class":"item"})
        
        # print(items)
        agent_banking_data[link]={}

        for item in items:
            paragraphs = item.find_all("p")
            list_items = item.find_all("li")
            tables = item.find_all("table")
            # print(paragraphs)
            header = item.a.text.strip()
            agent_banking_data[link][header]=[]

            child_divs = item.find_all(name='div',attrs={"class":"collapse accordion-answer"})
            # for child_div in child_divs:
            for paragraph in paragraphs:
                span = paragraph.find_all(name="span")
                if span:
                    agent_banking_data[link][header]=[paragraph.text.strip() for i in item.findAll("p")]
            for list_item in list_items:
                spans = list_item.find_all(name="span")
                # print(spans)
                if span:
                    agent_banking_data[link][header]=[i.text.strip() for i in item.findAll("span")]
            
            for table in tables:
                t_headers = []
                # Extract headers
                for th in table.find("thead").find_all("th"):
                    t_headers.append(th.text.strip())

                # Extract rows
                for tr in table.find("tbody").find_all("tr"):
                    row_data = {}
                    for td, t_header in zip(tr.find_all("td"), t_headers):
                        row_data[t_header] = td.text.strip()
                    agent_banking_data[link][header].append(row_data)
    except Exception as e:
        container=soup.find(name="div",attrs={"class": "container"})
        agent_banking_data[link]=container.getText()




agent_banking_data

```
