```python
import requests
from bs4 import BeautifulSoup


response=requests.get("https://citybankplc.com/")

data=response.text
```

```python
soup=BeautifulSoup(data,"html.parser")
```

```python
nav_bar=soup.find("div",{"class":"navigation-container"})
```

```python
hrefs_dict={}
for i in nav_bar.find_all(["li"]):
    # print(i.a.get_text())
    if i.a.text.strip()!="":
        text=i.a.text.replace("\n","")
        # sub_str="\r"
        # re = text.split(sub_str)
        # text=re[0]+sub_str
        # text=text.replace("\r","")
        print(text)
        # print(i)
        hrefs_dict[text]=[]
        try:
            for j in i.div.findAll(["a"],href=True):
                hrefs_dict[text].append(j["href"])
                print(j["href"])
                # print(j.a)
        except Exception as e:
            print(e)
        try:
            # for j in i.li.findAll(["a"],href=True):
            print(i)
            # print(list(i.descendants))
            print(i.find("a")["href"])
            hrefs_dict[text].append(i.find("a")["href"])
            # print(j["href"])
                # print(j.a)
        except Exception as e:
            print(e)
    print("#######"*25)
    print("===="*25)
    
```
