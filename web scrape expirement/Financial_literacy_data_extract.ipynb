```python
financial_literacy_data = {}

for link in list1:
    response = requests.get(link)
    data = response.text
    try:
        soup = BeautifulSoup(data, "html.parser")
        container = soup.find(name="div", attrs={"class": "question-container items"})
        items = container.find_all(name='div', attrs={"class": "item"})
        
        financial_literacy_data[link] = {}

        for item in items[:-2]:
            header = item.a.text.strip()
            financial_literacy_data[link][header] = []

            paragraphs = item.find_all("p")
            list_items = item.find_all("li")
            tables = item.find_all("table")
            child_divs = item.find_all(name='div', attrs={"class": "collapse accordion-answer"})

            # Store list items
            # for list_item in list_items:
            #     financial_literacy_data[link][header].append(list_item.text.strip())

            # Store paragraphs with spans
            for paragraph in paragraphs:
                # span = paragraph.find_all(name="span")
                # if span:
                    financial_literacy_data[link][header].append(paragraph.text.strip())

            # Check for Cluster Financing Policy link
            if header == "Self-Learning Material":
                link_tag = item.find("a", href=True)
                if link_tag:
                    financial_literacy_data[link][header].append(link_tag['href'])

    except Exception as e:
        container=soup.find(name="div",attrs={"class": "container"})
        financial_literacy_data[link]=container.getText()


financial_literacy_data
```

```python
import json
import os

# Specify the directory and filename
directory = 'D:\Internship\EnterpriseKnowledgeBased\data'  # Change this to your designated path
filename = 'financial_literacy_data.json'
file_path = os.path.join(directory, filename)

# Ensure the directory exists
os.makedirs(r'D:\Internship\EnterpriseKnowledgeBased\data', exist_ok=True)

# Write the JSON data to the file
with open(file_path, 'w') as f:
    json.dump(financial_literacy_data, f, indent=4)  # Using json.dump for better readability
```
