```python
import requests
from bs4 import BeautifulSoup
import json
```

```python
employee_banking_data = {}

for link in list1:
    response = requests.get(link)
    data = response.text
    try:
        # Parse the HTML content using BeautifulSoup
        soup = BeautifulSoup(data, "html.parser")
        
        # Locate the main container for the employee banking section
        container = soup.find(name="div", attrs={'class': "container"})
        
        if container:
            # Initialize a dictionary to store content from the current page
            content_dict = {}
            
            # Extract the description text
            description = container.find("p", align="justify").get_text(strip=True)
            # print(description)
            content_dict["description"] = description
            

            #Extract sections 
            sections = container.find_all(name='div',attrs={"class":"section"})


            # Extract Features
            features_section = sections[0]
            # print(features_section)

            if features_section:
                features = features_section.find_all("h5", class_="csrtitle")
                content_dict["Features"] = [feature.get_text(strip=True) for feature in features]
            
            # Extract Benefits
            benefits_section = sections[1]
            if benefits_section:
                benefit_lists = benefits_section.find_all("ul")
                benefit_headings = benefits_section.find_all("p")
                
                content_dict["Benefits"] = {
                    heading.find("strong").get_text(strip=True): [li.get_text(strip=True) for li in ul.find_all("li")]
                    for heading, ul in zip(benefit_headings, benefit_lists)
                }
            # Extract 'Why Bank with City' section
            why_bank_section =sections[2]
            if why_bank_section:
                why_bank_text = why_bank_section.find("p").get_text(strip=True)
                content_dict["why_bank_with_city"] = why_bank_text
            
            # Extract Award & Recognition section
            awards_section = sections[3]
            if awards_section:
                awards = awards_section.find_all("div", class_="text-section")
                content_dict["awards"] = [
                    {
                        "Title": award.find("h5", class_="title").get_text(strip=True),
                        "Detail": award.find("p", class_="detail").get_text(strip=True)
                    }
                    for award in awards
                ]
            
            # Add the content of the current page to the main dictionary
            employee_banking_data[link] = content_dict
        
    except Exception as e:
        print(f"An error occurred while processing {link}: {e}")


employee_banking_data
```

```python
import json
import os

# Specify the directory and filename
directory = 'D:\Internship\EnterpriseKnowledgeBased\data'  # Change this to your designated path
filename = 'employee_banking_data.json'
file_path = os.path.join(directory, filename)

# Ensure the directory exists
os.makedirs(r'D:\Internship\EnterpriseKnowledgeBased\data', exist_ok=True)

# Write the JSON data to the file
with open(file_path, 'w') as f:
    json.dump(employee_banking_data, f, indent=4)  # Using json.dump for better readability
```
