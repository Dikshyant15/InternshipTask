```python
cards_data={}


american_express_card_link="https://www.citybankplc.com/card/amex-cards"
cards_data[american_express_card_link]={}
response = requests.get(american_express_card_link)
data = response.text
try:
# Parse the page with BeautifulSoup
    soup = BeautifulSoup(data, "html.parser")

    # Find the container with all the cards
    card_container = soup.find(name='div', attrs={"class": "cardsfacecontainer"})

    # Find all the rows in the container
    card_rows = card_container.find_all(name='div', attrs={'class': 'row'})

    # Iterate over each row
    for row in card_rows:
        # Find all card boxes within the row
        card_boxes = row.find_all(name='div', attrs={'class': 'city-crcardbox'})
        
        # Extract all links from the row
        links = [a['href'] for a in row.find_all(name='a', href=True)]
        
        for i, card_box in enumerate(card_boxes):
            # Extract the card name
            card_name = card_box.find(name='h5', attrs={'class': 'title'}).text.strip()

            # Initialize the dictionary for the card
            cards_data[american_express_card_link][card_name] = {}

             # Add the link corresponding to the card
            if i < len(links):  # Ensure there's a link available for the card
                card_link = links[i]
                cards_data[american_express_card_link][card_name]['link'] = card_link
                
                # Fetch the content of the linked page
                card_response = requests.get(card_link)
                card_content = card_response.text
                
                # Parse the linked page content
                card_soup = BeautifulSoup(card_content, "html.parser")
                
                # # Extract specific content from the linked page (example: paragraphs)
                # # Adjust the selector based on the structure of the target pages
                # paragraphs = card_soup.find_all('p')
                # content = "\n".join([p.text.strip() for p in paragraphs])
                container= card_soup.find(name="div",attrs={"class":"question-container items"})
                items = container.findAll(name='div',attrs={"class":"item"})
                for item in items:
                            cards_data[american_express_card_link][card_name][item.a.text]=[]
                            cards_data[american_express_card_link][card_name][item.a.text]=[i.text for i in item.findAll("li")]
                # Store the content in the dictionary
                # cards_data[american_express_card_link][card_name]['content'] = content
except Exception as e:
        print(f"An error occurred while processing {american_express_card_link}: {e}")
cards_data
```
