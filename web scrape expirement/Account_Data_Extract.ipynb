```python
accounts_data={}
for parse_link in list1:
    accounts_data[parse_link]={}
    response=requests.get(parse_link)
    data=response.text
    try:
        soup=BeautifulSoup(data,"html.parser")
        faq=soup.find(name="div",attrs={"class": "question-container items"})
        obu=faq.findAll(name="div",attrs={"class":"item"})
        for a in obu:
            accounts_data[parse_link][a.a.text]=[]
            accounts_data[parse_link][a.a.text]=[i.text for i in a.findAll("li")]
    except Exception as e:
        faq=soup.find(name="div",attrs={"class": "container"})
        accounts_data[parse_link]=faq.getText()
        
        

accounts_data
```
