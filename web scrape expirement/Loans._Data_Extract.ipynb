```python
import requests
from bs4 import BeautifulSoup
import json
```

```python
with open(r"D:\Internship\EnterpriseKnowledgeBased\data\hrefs_dict2.json") as f:
    links_data=json.loads(f.read())
```

```python
list1=links_data.get("icon_loans\rLoansicon_arrow-front")
type(list1)
```

```python
list1
```

```python
loans_data = {}

for parse_link in list1:
    loans_data[parse_link] = {}
    response = requests.get(parse_link)
    data = response.text
    try:
        soup = BeautifulSoup(data,"html.parser")
        container= soup.find(name="div",attrs={"class":"question-container items"})
        items = container.findAll(name='div',attrs={"class":"item"})
        for item in items:
            loans_data[parse_link][item.a.text]=[]
            loans_data[parse_link][item.a.text]=[i.text for i in item.findAll("li")]
    except Exception as e:
        container=soup.find(name="div",attrs={"class": "container"})
        loans_data[parse_link]=container.getText()
loans_data
```

```python
import json
import os

# Specify the directory and filename
directory = 'D:\Internship\EnterpriseKnowledgeBased\data'  # Change this to your designated path
filename = 'loans_data.json'
file_path = os.path.join(directory, filename)

# Ensure the directory exists
os.makedirs(r'D:\Internship\EnterpriseKnowledgeBased\data', exist_ok=True)

# Write the JSON data to the file
with open(file_path, 'w') as f:
    json.dump(loans_data, f, indent=4)  # Using json.dump for better readability
```
